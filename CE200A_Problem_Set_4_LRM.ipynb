### Question 03

In an infinite one-dimensional reactor the diffusion coefficient is measured to be equal to 0.0025m$^2/s$. Consider these two scenarios:

(a) The initial concentration of an inert contaminant is 0.1 kg/m for $x < 10$ and zero for $x > 10$. Compute and plot the concentration profile at $t = 60$ s. (Use Python and include your code.)



```python

# Define the parameters
D = 0.0025 # [m^2/s]
C0 = 0.1 #[kg/m]
x0 = 10 #[m]
t = 60 # [s]

# Calculate the concentration profile
x = np.linspace(-0,20,100) 
C = 0.5*C0*(1-special.erf((x-x0)/np.sqrt(4*D*t)))

# Plot error function
fig , ax = plt.subplots(figsize=(7,7))
ax.plot(x, C, '.-') 
ax.set_xlabel('x [m]')
ax.set_ylabel('C [kg/m]')
ax.set_title('Problem 3a: Concentration at t = 60 s')
```

(b) The initial concentration is zero everywhere. Then, 1 kg of some chemical is deposited at $x=0$. If 60 s later 2 kg of the same chemical are added at $x= 2$ m, compute and plot the concentration profile at $t = 600$ s. (Use Python and include your code.)

```python
# Define the parameters
D = 0.0025 # [m^2/s]
C0 = 0.1 #[kg/m]
M1 = 1 # [kg]
M2 = 2 # [kg]
x01 = 0 #[m]
x02=  2 # (m)
t1 = 600 # [s]
t2 = 540 # [s]
# Calculate the concentration profile
x = np.linspace(-20,20,100) 
C1 = (M1/np.sqrt(4*D*t1))*np.exp(-(x-x01)**2/np.sqrt(4*D*(t1)))
C2 = (M2/np.sqrt(4*D*t1))*np.exp(-(x-x02)**2/np.sqrt(4*D*(t2)))
C3=C1+C2
# Plot error function
fig , ax = plt.subplots(figsize=(6,6))
ax.plot(x, C1, '.-', label='C1') 
ax.plot(x, C2, '.-',label='C2') 
ax.plot(x, C3, '.-',label='C3') 
ax.legend()
ax.set_xlabel('x [m]')
ax.set_ylabel('C [kg/m]')
ax.set_title('Problem 3b: Concentration at t = 600 s')
```

(b) Plot your solution with Python (include your code) for $\theta_0 = 10 K$, $ω = 2\pi \times 10^{−5}$ s$^{-1}$, $\kappa = 0.14 \times 10^{−6}$ m$^2$/s for $t = 0,6,12,18,24$ hr for the first 1 m of the water column. [See the instructions on how to publish your Python code.]


Define parameters:

```python
theta0 =10 #[K]
w= 2*np.pi/100000.  #[s^-1]
k= 0.14/1000000  # [m2/s]
```

Plot the solution at different times (all on the same plot):


```python
# Define the time array
t = np.array([0, 6, 12, 18, 24]) #[hr]
t = t*3600 #Convert to seconds

# Define the z coordinate
z = np.linspace(0,-1,100) #[m]

# Define the temperature profile at each time
theta = np.zeros([z.shape[0],t.shape[0]])
for ii in range(t.shape[0]): #loop over the different times
  theta[:,ii] = theta0*np.exp(np.sqrt(w/(2*k))*z)*np.cos(w*t[ii]+np.sqrt(w/(2*k))*z)

# Plot the solution
fig, ax = plt.subplots()
ax.plot(theta,z) #this will plot all times on one plot

#set title, xlabel, ylabel here
ax.legend(['0 hr','6 hr', '12 hr', '18 hr', '24 hr']) #make sure this shows up correctly for your plot
ax.set_xlabel('Theta [K]')
ax.set_ylabel('Z [m]')
ax.set_title('Problem 4: Heat diffusion in lake')
```
